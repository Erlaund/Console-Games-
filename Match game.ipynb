```python
import time
import math

print('Спички - не игрушка, но если очень хочется, то можете поиграть в эту игру')
time.sleep(2) #Пауза 
print()

text = '''
==== Правила игры ====\n
У вас есть опредленное количество спичек и вы можете задать лимит взятия спичек. 
2 игрока забирают эти спички поочереди, не первышая лимит.
Выигрывает тот, кто возьмет последнюю спичку.
'''

player = 'компьютер'
winner = 'компьютер'
i = 1

print(text)
time.sleep(3) #Пауза 
print("И кстати, для упрощения, спички будут визуально отображаться до 10 штук в одной строке")
print("Удачной игры!")
print()
print("P.S. Не ставьте количество спичек больше 100 и лимит взятия больше 10.")
print("Компьютер, конечно, может с вами поиграть и при других вводных, однако устанете вы раньше, чем выиграете")

time.sleep(4) #Пауза 
print()
print()

# Настройки игры

print("Введите количество спичек в игре, но не больше 100")
n = int(input())
print("Введите лимит взятия спичек, но не больше 10")
k = int(input())

print("Кто начинает игру? Компьютер или игрок?")
game = str(input())

if game == "Компьютер":
    start = 1
    print('Игру начинает компьютер')
    print()
elif game == "Игрок":
    start = 2
    print('Игру начинаете вы')
    print()
else:
    start = 0
    print('Итак со спичками балуетесь. Хотя бы с вводом не играйте')
        

def match_print(n): # функция визуализация спичек
    cost1_n = n//10
    cost2_n = n%10
    
    for i in range(cost1_n):
        print(str(i+1) + '. '+'||| '* (10 // 3) + '|' * (10 % 3))
    
    if cost2_n >0:
        print(str(cost1_n+1) + '. ' + '||| '* (cost2_n // 3) + '|' * (cost2_n % 3) )
    return " " 

# Ходы компьютера и игрока

if start == 1 or start == 2:
    while n>0:
        if start == 1: # Ход компьютера
            print('Ход №', i)
            print(match_print(n))
            # a = int(math.fabs(n%k - 1)) # функция беспроигрышной игры
            a = n - ((n-1) // k * k)  # функция для обычной игры
            if a<=0:
                a = a + 1
            elif k>=n:
                a=n
            print('Компьютер берет', a)
            print()
            n = n - a 
            start = 2
            i+=1
            time.sleep(1.5)  #Пауза между ходами

        else: # Ход игрока
            print('Ход №', i)
            print(match_print(n))
            a = 0
            while a <= 0 or a > k:
                a = int(input('Сколько спичек? '))
                if a <= 0 or a > k:
                    print('Ну и что вы вводите? Прочитайте правила еще раз, если непонятно')
            print()
            n = n - a
            start = 1
            i+=1
            time.sleep(1.5)  #Пауза между ходами
            if n <= 0: # Условный оператор для определения победителя
                winner = 'игрок'
            player = 'компьютер'
        
           
print()
print('победил ',winner)
```
